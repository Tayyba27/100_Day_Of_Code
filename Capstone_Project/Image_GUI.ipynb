####  Using what you have learnt about Tkinter, you will create a desktop application with a Graphical User Interface (GUI) where you can upload an image and use Python to add a watermark logo/text.

```python
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
from PIL import Image, ImageDraw, ImageFont

# Function to upload an image
def upload_image():
    global img, img_path
    img_path = filedialog.askopenfilename(title="Select an Image", filetypes=[("Image files", "*.png;*.jpg;*.jpeg;*.bmp;*.gif")])
    
    if img_path:
        img = Image.open(img_path)
        messagebox.showinfo("Image Upload", "Image loaded successfully!")

# Function to add watermark text
def add_watermark_text():
    if not img:
        messagebox.showerror("Error", "Please upload an image first!")
        return

    watermark_text = watermark_entry.get()
    if not watermark_text:
        messagebox.showerror("Error", "Please enter watermark text!")
        return

    # Add text watermark
    width, height = img.size
    draw = ImageDraw.Draw(img)
    font = ImageFont.load_default()
    text_width, text_height = draw.textsize(watermark_text, font=font)
    
    # Positioning the text at the bottom-right corner
    position = (width - text_width - 10, height - text_height - 10)
    draw.text(position, watermark_text, font=font, fill=(255, 255, 255, 128))  # white text with some transparency
    
    messagebox.showinfo("Watermark Added", "Text watermark added successfully!")

# Function to add watermark logo
def add_watermark_logo():
    if not img:
        messagebox.showerror("Error", "Please upload an image first!")
        return
    
    logo_path = filedialog.askopenfilename(title="Select a Watermark Logo", filetypes=[("Image files", "*.png;*.jpg;*.jpeg;*.bmp;*.gif")])
    if not logo_path:
        return

    logo = Image.open(logo_path).convert("RGBA")
    img_width, img_height = img.size
    logo_width, logo_height = logo.size

    # Positioning the logo at the bottom-right corner
    position = (img_width - logo_width - 10, img_height - logo_height - 10)

    # Paste the logo on the image
    img.paste(logo, position, logo)  # The third argument is the alpha mask for transparency
    messagebox.showinfo("Watermark Added", "Logo watermark added successfully!")

# Function to save the edited image
def save_image():
    if not img:
        messagebox.showerror("Error", "Please upload an image first!")
        return

    save_path = filedialog.asksaveasfilename(defaultextension=".png", filetypes=[("PNG files", "*.png"), ("JPEG files", "*.jpg"), ("All files", "*.*")])
    if save_path:
        img.save(save_path)
        messagebox.showinfo("Image Saved", "Image saved successfully!")

# Main window setup
root = tk.Tk()
root.title("Watermark Image Editor")
root.geometry("500x400")

# Variable to hold the uploaded image
img = None
img_path = ""

# Add UI elements
upload_button = tk.Button(root, text="Upload Image", command=upload_image)
upload_button.pack(pady=10)

watermark_label = tk.Label(root, text="Enter watermark text:")
watermark_label.pack(pady=5)

watermark_entry = tk.Entry(root, width=40)
watermark_entry.pack(pady=5)

text_watermark_button = tk.Button(root, text="Add Text Watermark", command=add_watermark_text)
text_watermark_button.pack(pady=10)

logo_watermark_button = tk.Button(root, text="Add Logo Watermark", command=add_watermark_logo)
logo_watermark_button.pack(pady=10)

save_button = tk.Button(root, text="Save Image", command=save_image)
save_button.pack(pady=20)

# Run the Tkinter event loop
root.mainloop()
```
