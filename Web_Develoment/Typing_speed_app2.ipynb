```python
import tkinter as tk
import time
import random

# Sample texts for typing test
sample_texts = [
    "The quick brown fox jumps over the lazy dog.",
    "Python is a powerful and versatile programming language.",
    "Tkinter makes it easy to build desktop applications with Python."
]

class TypingSpeedTestApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Typing Speed Test")
        
        self.text = random.choice(sample_texts)
        self.start_time = None
        
        self.label = tk.Label(root, text="Type the text below:", font=("Arial", 14))
        self.label.pack(pady=10)
        
        self.text_label = tk.Label(root, text=self.text, wraplength=400, font=("Arial", 12), justify="center")
        self.text_label.pack(pady=10)
        
        self.entry = tk.Entry(root, width=50, font=("Arial", 12))
        self.entry.pack(pady=10)
        self.entry.bind("<KeyPress>", self.start_timer)
        
        self.button = tk.Button(root, text="Submit", command=self.calculate_speed)
        self.button.pack(pady=10)
        
        self.result_label = tk.Label(root, text="", font=("Arial", 12))
        self.result_label.pack(pady=10)

        self.reset_button = tk.Button(root, text="Restart", command=self.reset_test)
        self.reset_button.pack(pady=5)

    def start_timer(self, event):
        if self.start_time is None:  # Start the timer on the first keypress
            self.start_time = time.time()

    def calculate_speed(self):
        if self.start_time is None:
            self.result_label.config(text="Start typing first!")
            return
        
        end_time = time.time()
        typed_text = self.entry.get()
        
        words_typed = len(typed_text.split())
        time_taken = (end_time - self.start_time) / 60  # Convert to minutes
        
        if time_taken > 0:
            wpm = words_typed / time_taken
            accuracy = self.calculate_accuracy(typed_text)
            self.result_label.config(text=f"Your typing speed: {wpm:.2f} WPM\nAccuracy: {accuracy:.2f}%")
        else:
            self.result_label.config(text="Error in time calculation.")
        
        self.start_time = None  # Reset timer

    def calculate_accuracy(self, typed_text):
        """Calculate the accuracy based on the percentage of matching words."""
        typed_words = typed_text.split()
        original_words = self.text.split()
        correct_words = sum(1 for tw, ow in zip(typed_words, original_words) if tw == ow)
        total_words = max(len(original_words), 1)
        return (correct_words / total_words) * 100

    def reset_test(self):
        """Reset the test with a new random text and clear the entry field."""
        self.text = random.choice(sample_texts)
        self.text_label.config(text=self.text)
        self.entry.delete(0, tk.END)
        self.result_label.config(text="")
        self.start_time = None

if __name__ == "__main__":
    root = tk.Tk()
    root.geometry("500x400")
    app = TypingSpeedTestApp(root)
    root.mainloop()
```
